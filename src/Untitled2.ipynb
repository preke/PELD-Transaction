```python
import pandas as pd
import numpy as np

df_peld = pd.read_csv('../Dyadic_PELD.tsv', sep='\t')
print(df_peld.shape)
df_peld.head()
```

```python
Mood_dict = {
    'neutral': [0.0, 0.0, 0.0],
    'M1' : [0.605, 0.59 , 0.165],
    'M2': [-0.57666667,  0.58666667, -0.02333333],
    'M3': [-0.63, -0.27, -0.33]
}


def emo_to_mood(emotion):
    if emotion == 'neutral': return Mood_dict['neutral']
    elif emotion == 'joy' or emotion == 'surprise': return Mood_dict['M1']
    elif emotion == 'sadness': return Mood_dict['M3']
    else: return Mood_dict['M2']

df_peld['Mood_1'] = df_peld['Emotion_1'].apply(emo_to_mood)
df_peld['Mood_2'] = df_peld['Emotion_2'].apply(emo_to_mood)
df_peld['Mood_3'] = df_peld['Emotion_3'].apply(emo_to_mood)

df_peld.head()
```

```python
Emotion_dict = {
    'anger': [-0.51, 0.59, 0.25],
    'disgust': [-0.60, 0.35, 0.11],
    'fear': [-0.62, 0.82, -0.43],
    'joy': [0.81, 0.51, 0.46],
    'neutral': [0.0, 0.0, 0.0],
    'sadness': [-0.63, -0.27, -0.33],
    'surprise': [0.40, 0.67, -0.13]
}


def get_tran_mood(personality, mood, emotion):
    return [0.9*(eval(personality)[4] + 1.0)*i for i in Emotion_dict[emotion]]

df_peld['tran_mood'] = df_peld.apply(lambda r: get_tran_mood(r['Personality'], r['Mood_1'], r['Emotion_2']), axis=1)
```

```python

from sklearn.metrics import mean_squared_error

df = df_peld[df_peld['Personality'] == '[0.635, 0.354, 0.521, 0.552, 0.469]']

result = 0.0
for r in df.iterrows():
    result += mean_squared_error(r[1]['tran_mood'], r[1]['Mood_3'])
print(result/len(df))

result = 0.0
for r in df.iterrows():
    result += mean_squared_error(r[1]['Mood_1'], r[1]['Mood_3'])
print(result/len(df))

result = 0.0
for r in df.iterrows():
    result += mean_squared_error(r[1]['Mood_2'], r[1]['Mood_3'])
print(result/len(df))


result = 0.0
for r in df.iterrows():
    result += mean_squared_error(Emotion_dict[r[1]['Emotion_1']], r[1]['Mood_3'])
print(result/len(df))

result = 0.0
for r in df.iterrows():
    result += mean_squared_error(Emotion_dict[r[1]['Emotion_2']], r[1]['Mood_3'])
print(result/len(df))

result = 0.0
for r in df.iterrows():
    result += mean_squared_error(Emotion_dict[r[1]['Emotion_3']], r[1]['Mood_3'])
print(result/len(df))
```
